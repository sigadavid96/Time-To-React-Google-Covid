## Initialize TrendReq
To make pytrends API requests, we create a `TrendReq` from `pytrends.request` library. Here, we also add a backoff_factor to include a backoff when doing retries on the API


```python
pytrend = TrendReq(backoff_factor=0.1)                
#A backoff factor to apply between attempts after the second try
```

## Get historical data by region
The get_historical_data_by_region() function with keyword list,  region and sleep parameters provides historical data for given keyword and region. In our case, we download only historical data from 10/2019 to 05/2022


The interest values are calculated on a scale from 0 to 100, where 100 is the search with the most popularity as a fraction of total searches in that location. A value of 0 indicates a location where there was not enough data for this term. Source → [Google Trends](https://support.google.com/trends/answer/4355212)..

```python
def get_historical_data_by_region(kw_list,region,sleep=0):
  historicaldf = pytrend.get_historical_interest(kw_list, year_start=2019, month_start=10, day_start=1, hour_start=0, year_end=2022, month_end=5, day_end=1, hour_end=0, cat=0, geo=region, gprop='', sleep=sleep)
  return historicaldf
```

We now build the payload for our historical data pull. In this case we use `Covid` and `Coronavirus` as our keywords


```python
kw_list=['Covid', 'Coronavirus']
pytrend.build_payload(kw_list, timeframe='today 1-m')
```

We can use pytrend.related_queries() to get list of queries related to given keyword (in our case Covid, Coronavirus).

```python
related_queries = pytrend.related_queries()
related_queries.values()
```

``df.to_csv`` saves the file as raw csv file as colab. We can then store this in github

```python
def get_data_for_region(region):
  df = get_historical_data_by_region(kw_list,region,sleep=20)
  #Adding region column to the dataframe
  df['Region'] = region
  df = df.drop(columns='isPartial')
  #Save the data to csv
  csv = df.to_csv('au_data.csv', index=True)
  return df
```

# Data Processing



We process this data for each country i.e we group by data, set index of df to `date`, combine our two query terms (`Covid` and `Coronavirus`), and get the 7 day rolling mean

```python
# Resetting the index to date, it helps with plotting
def group_by_date(google_trends):
  times = pd.to_datetime(google_trends.date)
  tempDf=pd.DataFrame(google_trends.groupby([times.dt.date]).Covid.mean())
  corona=google_trends.groupby([times.dt.date]).Coronavirus.mean()
  google_trends=tempDf
  google_trends['Coronavirus']=corona
  google_trends.index=pd.to_datetime(google_trends.index)
  return google_trends
```

```python
# We have two terms and we combine it to get our trends data processing
def combine_query_terms(google_trends):
  google_trends['covid_trend'] = google_trends['Covid'] + google_trends['Coronavirus']
  return google_trends
```

```python
ROLLING_MEAN_DAYS = 7
def get_rolling_mean_trends(google_trends):
  # Rolling mean data
  rolling_google_trends=pd.DataFrame(google_trends.covid_trend.rolling(ROLLING_MEAN_DAYS, min_periods=1).mean())
  rolling_google_trends.index=pd.to_datetime(rolling_google_trends.index)
  return rolling_google_trends
```

```python
def process_pytrends_data(google_trends):
  google_trends = group_by_date(google_trends)
  google_trends = combine_query_terms(google_trends)
  rolling_mean = get_rolling_mean_trends(google_trends)
  return google_trends, rolling_mean
```

```python
google_us_trends, rolling_google_us_trends = process_pytrends_data(google_us_trends)
google_gb_trends, rolling_google_gb_trends = process_pytrends_data(google_gb_trends)
google_in_trends, rolling_google_in_trends = process_pytrends_data(google_in_trends)
google_au_trends, rolling_google_au_trends = process_pytrends_data(google_au_trends)
google_it_trends, rolling_google_it_trends = process_pytrends_data(google_it_trends)
```

## Getting Covid Statistics

We now get the covid statistics to plot. `CovidStatistics` holds the covid statistics chosen from our feature selection

```python
def get_covid_statistics(country):
  df_pos_rate_nonzero, df_pos_rate_full_nonzero = get_subset_owid_data(country, 'positive_rate')
  df_new_test_nonzero, df_new_tests_full_nonzero = get_subset_owid_data(country, 'new_tests_smoothed_per_thousand')
  df_death_rate_nonzero, df_death_rate_full_nonzero = get_subset_owid_data(country, 'new_deaths_smoothed_per_million')
  df_new_cases_nonzero, df_new_cases_full_nonzero = get_subset_owid_data(country, 'new_cases_smoothed_per_million')
  print(df_new_cases_nonzero)
  return df_pos_rate_nonzero, df_pos_rate_full_nonzero, df_new_test_nonzero, df_new_tests_full_nonzero, df_death_rate_nonzero, df_death_rate_full_nonzero, df_new_cases_nonzero, df_new_cases_full_nonzero
```

```python
def covid_metric_and_google(df_pos_rate_nonzero, rolling_google_trends, metric):
  Covid_Positive_And_Google=pd.merge(df_pos_rate_nonzero, rolling_google_trends, left_index=True, right_index=True)
  # Covid_Positive_And_Google.describe()
  scaler = MinMaxScaler(feature_range=(0, 100))
  Covid_Positive_And_Google[[metric,'covid_trend']] = scaler.fit_transform(Covid_Positive_And_Google[[metric,'covid_trend']])
  Covid_Positive_And_Google.index = pd.to_datetime(Covid_Positive_And_Google.date)
  return Covid_Positive_And_Google
# Scaling by 100
```

```python
class CovidStatistics:
  # Get all the related plots for a country 
    def __init__(self, country):
      self.df_pos_rate_nonzero, self.df_pos_rate_full_nonzero = get_subset_owid_data(country, 'positive_rate')
      self.df_new_test_nonzero, self.df_new_tests_full_nonzero = get_subset_owid_data(country, 'new_tests_smoothed_per_thousand')
      self.df_death_rate_nonzero, self.df_death_rate_full_nonzero = get_subset_owid_data(country, 'new_deaths_smoothed_per_million')
      self.df_new_cases_nonzero, self.df_new_cases_full_nonzero = get_subset_owid_data(country, 'new_cases_smoothed_per_million')

    # Plot the covid Statistics 
    def plot_statistics(self):
      self.df_pos_rate_nonzero.plot()
      self.df_new_test_nonzero.plot()
      self.df_death_rate_nonzero.plot()
      self.df_new_cases_nonzero.plot()
      pyplot.xticks(rotation = 45)
      pyplot.show()

    # Create metrics based off the rolling mean 
    def create_metrics(self, rolling_mean):
      self.Covid_Positive_Rate_And_Google = covid_metric_and_google(self.df_pos_rate_nonzero, rolling_mean, 'positive_rate')
      self.Covid_Tests_And_Google = covid_metric_and_google(self.df_new_test_nonzero, rolling_mean, 'new_tests_smoothed_per_thousand')
      self.Covid_Deaths_And_Google = covid_metric_and_google(self.df_death_rate_nonzero, rolling_mean, 'new_deaths_smoothed_per_million')
      self.Covid_Cases_And_Google = covid_metric_and_google(self.df_new_cases_nonzero, rolling_mean, 'new_cases_smoothed_per_million')
```

## Data Transformation for Modelling

```python
# This code is to check data with some "per" % missing values and drop those columns

def drop_cols_threshold(df,per=75,drop_object=True):
  # ref : https://stackoverflow.com/questions/47414848/pandas-select-all-columns-without-nan
  cols_to_delete = df.columns[df.isnull().sum()/len(df) > (per/100)]
  df.drop(cols_to_delete, axis = 1, inplace = True)
  if drop_object==True:
     df=df.select_dtypes(exclude=['object'])
  return df
```

```python
# This code is to clean, scale and select good quality columns for modelling exercise
def transform_data_for_modelling(covid_stats, rolling_mean):
  full_Covid_Positive_And_Google = pd.merge(covid_stats.df_death_rate_full_nonzero, rolling_mean, left_index=True, right_index=True)
  # Allowing for only 5% missingness in data, and dropping object columns
  full_Covid_Positive_And_Google=drop_cols_threshold(full_Covid_Positive_And_Google,5)
  scaler = MinMaxScaler(feature_range=(0, 100))
  scaler.fit(full_Covid_Positive_And_Google)
  df_scaled=scaler.fit_transform(full_Covid_Positive_And_Google)
  scale_full_CPAG=pd.DataFrame(df_scaled, columns=list(full_Covid_Positive_And_Google.columns))
  return full_Covid_Positive_And_Google, scale_full_CPAG
```

# Time to React Analysis

## Covid Positivity Data : Finding the Critical Points (minimas and maximas) 

```python
def get_minnimas_maximas(df,var,width=10):
  df['min'] = df[var][(df[var].shift(width) > df[var]) & (df[var].shift(-1*width) > df[var])]
  df['max'] = df[var][(df[var].shift(width) < df[var]) & (df[var].shift(-1*width) < df[var])]
  return df
```

```python
def get_firstpeak(Covid_Positive_And_Google,var_of_int,width=10):
  # Building an algorithm to identify critical points
  covid_df=get_minnimas_maximas(Covid_Positive_And_Google,var_of_int,width)
  # Plot results
  group_df_maxdates=covid_df[covid_df['max'].notnull()]
  group_df_maxdates['date'] = pd.to_datetime(group_df_maxdates['date'])
  group_df_maxdates['months']=group_df_maxdates['date'].dt.to_period('M')
  # Picking the last maxima to identify the critical point which is about to Shoot! 
  group_df_maxdates=group_df_maxdates.groupby('months').first()
  first_covid_peak=group_df_maxdates.iloc[0]['date']
  # print(group_df_maxdates.iloc[0]['date'])
  return first_covid_peak
```

```python
def get_interest_gap(stat, metric, width_param=10):
  covid_peak=get_firstpeak(stat,'covid_trend',width_param)
  interest_peak=get_firstpeak(stat, metric,width_param)
  return interest_peak-covid_peak
```

```python
def get_interest_gap_by_country(stats, width_param=10):
  print("Positive Rate", get_interest_gap(stats.Covid_Positive_Rate_And_Google, "positive_rate"))
  print("New Deaths", get_interest_gap(stats.Covid_Deaths_And_Google, "new_deaths_smoothed_per_million"))
  print("New Tests", get_interest_gap(stats.Covid_Tests_And_Google, "new_tests_smoothed_per_thousand"))
  print("New Cases", get_interest_gap(stats.Covid_Cases_And_Google, "new_cases_smoothed_per_million"))
```

We use the maxima and minimas to get the delay in covid search trends vs the increase in factors above

## Time Series Forecasting

Here we try to use the pytrends data to analyze if there is a pattern in trend searches and forecast for future dates. 

```python
#forecast covid data
import matplotlib.pyplot as plt

#Function to split data into train and test data and plot the trend
def forecasting_trends(df):
  #Split the data into train and test data
  train = pd.DataFrame(df[df.index < pd.to_datetime("2022-03-15", format='%Y-%m-%d')])
  test = pd.DataFrame(df[df.index > pd.to_datetime("2022-03-15", format='%Y-%m-%d')])

  #Visualize the output
  plt.figure(figsize=(25, 8))
  plt.plot(train, color = "black")
  plt.plot(test, color = "red")
  plt.ylabel('Coronovirus')
  plt.xlabel('Date')
  plt.xticks(rotation=45)
  plt.title("Train/Test split for Covid Data")
  plt.show()
  return train, test
```

```python
df = google_us_trends['covid_trend']
train, test = forecasting_trends(df)
```

As there are huge variations in the trend, we have to smoothen the data before fitting the model. Lets try smoothening using rolling mean of 7 days

```python
#lets smoothen the data using rolling mean of 7 days
plt.figure(figsize=(25, 8))
rolmean = df.rolling(7).mean()
rolstd = df.rolling(7).std()
plt.plot(df)
plt.plot(rolmean)
plt.plot(rolstd)
plt.show()
```

```python
#Using rolling mean data for forecasting
import matplotlib.pyplot as plt
df = rolmean
train, test = forecasting_trends(df)
```

```python
!pip install statsmodels==0.12.1
import statsmodels
#Import libraries for time series forecasting
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
#Using Arima model for forecasting

plt.figure(figsize = (25,8))
plt.plot(train, color = "black")
plt.plot(test, color = "red")
plt.ylabel('Coronovirus')
plt.xlabel('Date')
plt.xticks(rotation=45)

#Get the interest metric
train.dropna(inplace=True)
y = train['covid_trend']

#Fit the model
ARIMAmodel = ARIMA(y, order = (2, 2, 2))
ARIMAmodel = ARIMAmodel.fit()

#Forecast
y_pred = ARIMAmodel.get_forecast(len(test.index))
y_pred_df = y_pred.conf_int(alpha = 0.05) 
y_pred_df["Predictions"] = ARIMAmodel.predict(start = y_pred_df.index[0], end = y_pred_df.index[-1])
y_pred_df.index = test.index
y_pred_out = y_pred_df["Predictions"] 

#Plot the results
plt.plot(y_pred_out, color='Green', label = 'ARIMA Predictions')
plt.legend()

import numpy as np
from sklearn.metrics import mean_squared_error

#Evaluation of forecast
arma_rmse = np.sqrt(mean_squared_error(test['covid_trend'].values, y_pred_df["Predictions"]))
print("RMSE: ",arma_rmse)
```

This forecasting relies only on the previous search interest parameter. For events like covid, additional features like vaccination rate, government measures, potential threat from new variants should be used to forecast accurately. Since there was no specific pattern in the train data, the forecast was not accurate to the test data. However, in scenarios where the trend follows a specific cycle (like annual events) then forecasting would have a good scope to predict the trend.
